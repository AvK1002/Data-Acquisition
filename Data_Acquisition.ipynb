# Identify any potential outliers in a numerical column of your choice. Explain your approach.


1. Find the first quartile (Q1), which represents the 25th percentile of the data.

2. Find the third quartile (Q3), which represents the 75th percentile of the data.
3.Calculate IQR:
Calculate the IQR as the difference between Q3 and Q1 (IQR = Q3 - Q1).
4. Determine Bounds:
Calculate the lower bound as Q1 - 1.5 * IQR.
Calculate the upper bound as Q3 + 1.5 * IQR.
Identify Outliers:

 Any data point that falls below the lower bound or above the upper bound is considered a potential outlier.

```python
import matplotlib.pyplot as plt
Q1_hia = df['Household Internet Access (%)'].quantile(0.25)
Q3_hia = df['Household Internet Access (%)'].quantile(0.75)
IQR_hia = Q3_hia - Q1_hia
lower_bound_hia = Q1_hia - 1.5 * IQR_hia
upper_bound_hia = Q3_hia + 1.5 * IQR_hia

outliers_hia = df[(df['Household Internet Access (%)'] < lower_bound_hia) | (df['Household Internet Access (%)'] > upper_bound_hia)]
print("Potential outliers in Household Internet Access (%):")
if outliers_hia.empty:
    print("No outliers found.")
else:

    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df['Household Internet Access (%)'])
    plt.title('Box Plot of Household Internet Access (%)')
    plt.show()
    print(outliers_hia)
```
